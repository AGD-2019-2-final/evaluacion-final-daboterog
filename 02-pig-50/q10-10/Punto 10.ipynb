```python
%%pig
data = LOAD 'data.csv' USING PigStorage(',')
    AS (f1:INT, f2:CHARARRAY, f3:CHARARRAY, f4:CHARARRAY, f5:CHARARRAY, f6:INT);
DUMP data;
```

```python
%%pig
a = FOREACH data GENERATE f3, SIZE(f3);
DUMP a;
```

```python
%%pig
b = ORDER a BY $1 DESC, $0;
DUMP b;
```

```python
%%pig
c = LIMIT b 5;
DUMP c;
```

```python
%%pig
STORE c INTO 'output';
fs -get output/ .
```
