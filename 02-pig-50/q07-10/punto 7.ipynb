```python
%%pig

data = LOAD 'data.tsv'
    AS (f1:CHARARRAY, f2:bag{}, f3:MAP[]);
DUMP data;
```

```python
%%pig
a = FOREACH data GENERATE f1, COUNT(f2), COUNT(KEYSET(f3));
DUMP a;
```

```python
%%pig
b = ORDER a BY $0, $1, $2;
DUMP b;
```

```python
%%pig
STORE b INTO 'output';
fs -get output/ .
```
