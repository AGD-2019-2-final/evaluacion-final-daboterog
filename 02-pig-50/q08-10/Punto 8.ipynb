```python
%%pig

data = LOAD 'data.tsv'
    AS (f1:CHARARRAY, f2:bag{}, f3:MAP[]);
DUMP data;
```

```python
%%pig
a = FOREACH data GENERATE FLATTEN(f2), FLATTEN(KEYSET(f3));
DUMP a;
```

```python
%%pig
b = FOREACH a GENERATE ($0, $1) AS c;
DUMP b;
```

```python
%%pig
d = GROUP b BY c;
DUMP d;
```

```python
%%pig
e = FOREACH d GENERATE group, COUNT(b);
DUMP e;
```

```python
%%pig
STORE e INTO 'output';
fs -get output/ .
```
